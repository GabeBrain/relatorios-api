# ***DETECÇÃO DE OUTLIERS NA BASE***

```python
# --- Detecção de Outliers - Utilizando o invervalo interquartil ---
print("\n\n--- Detecção de Outliers ---")
columns_to_check = [
    #'Quartos', 'Garagem', 'Torres', 'Elevadores', 'Unidades por Tipologia',
    'M2 Privativo',
    #'Oferta Lançada', 'Taxa AdministrativaEntrada', 'Nº de Parcelas',
    #'Valor das parcelas', '% de Juros Mensal', 'Indíce de Juros', 'Desconto à Vista',
    #'Tempo de Venda', 'Quantidade de Suítes', 'Banheiros', 'Estoque', 'Vendas',
    'Preço',
    #'VGV Total', 'VGV Oferta Final', 'Preço de Lançamento'
]

outliers_report = {}
outliers_by_padrao = {} # Dicionário para armazenar DataFrames de outliers por Padrão

if 'Padrão' in df.columns:
    for padrao in df['Padrão'].unique():
        df_padrao = df[df['Padrão'] == padrao].copy() # Create a copy to avoid SettingWithCopyWarning
        print(f"\nAnalisando outliers para o Padrão: {padrao}")
        padrao_outliers_list = [] # List to collect outliers for the current Padrão
        total_rows_padrao = len(df_padrao) # Total rows for the current Padrão

        for col in columns_to_check:
            if col in df_padrao.columns and pd.api.types.is_numeric_dtype(df_padrao[col]):
                Q1 = df_padrao[col].quantile(0.25)
                Q3 = df_padrao[col].quantile(0.75)
                IQR = Q3 - Q1
                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR

                outliers = df_padrao[(df_padrao[col] < lower_bound) | (df_padrao[col] > upper_bound)].copy()
                if not outliers.empty:
                    outliers['Tipo_Erro'] = f'Outlier_{col}_{padrao}'
                    padrao_outliers_list.append(outliers)
                    # Calculate and store percentage in report
                    outlier_count = len(outliers)
                    percentage = (outlier_count / total_rows_padrao) * 100 if total_rows_padrao > 0 else 0
                    if padrao not in outliers_report:
                        outliers_report[padrao] = {}
                    outliers_report[padrao][col] = {"count": outlier_count, "percentage": percentage}


        if padrao_outliers_list:
            outliers_by_padrao[padrao] = pd.concat(padrao_outliers_list)

else:
    print("Coluna 'Padrão' não encontrada. A detecção de outliers será feita na base completa.")
    all_outliers_list = []
    total_rows_base = len(df) # Total rows for the complete base
    for col in columns_to_check:
        if col in df.columns and pd.api.types.is_numeric_dtype(df[col]):
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].copy()
            if not outliers.empty:
                outliers['Tipo_Erro'] = f'Outlier_{col}'
                all_outliers_list.append(outliers)
                # Calculate and store percentage in report
                outlier_count = len(outliers)
                percentage = (outlier_count / total_rows_base) * 100 if total_rows_base > 0 else 0
                if 'Base Completa' not in outliers_report:
                     outliers_report['Base Completa'] = {}
                outliers_report['Base Completa'][col] = {"count": outlier_count, "percentage": percentage}

    if all_outliers_list:
        outliers_by_padrao['Base Completa'] = pd.concat(all_outliers_list)


if outliers_by_padrao:
    print("\nResumo de Outliers encontrados por Padrão e Coluna:")
    for padrao, report_data in outliers_report.items():
        print(f"\n--- Padrão: {padrao} ---")
        for col, data in report_data.items():
             print(f"  Outlier_{col}_{padrao}: {data['count']} outliers ({data['percentage']:.2f}%)")


    # Exporta outliers para Excel com planilhas separadas por Padrão
    output_file_name = "Outliers_Por_Padrao.xlsx"
    with pd.ExcelWriter(output_file_name) as writer:
        for padrao, outliers_df in outliers_by_padrao.items():
            outliers_df.to_excel(writer, sheet_name=padrao, index=False)

    print(f"\nAs linhas com outliers foram exportadas para: [1m{output_file_name}[0m em planilhas separadas por Padrão.")
    files.download(output_file_name)

else:
    print("\nNenhum outlier significativo encontrado nas colunas especificadas.")
```

# ***VALIDAÇÃO DE DATAS DE ENTREGA - EXPORTAR ARQUIVO COM OS ERROS***

```python
# --- Validação de Datas ---
print("\n\n--- Validação de Consistência de Datas ---")
if 'Data de Entrega' in df.columns and 'Data de Lançamento' in df.columns:
    # Converte as colunas para o formato de data, tratando erros
    df['Data de Entrega'] = pd.to_datetime(df['Data de Entrega'], errors='coerce')
    df['Data de Lançamento'] = pd.to_datetime(df['Data de Lançamento'], errors='coerce')

    # Identifica linhas onde a data de entrega é anterior à de lançamento
    inconsistent_dates = df[df['Data de Entrega'] < df['Data de Lançamento']]

    if not inconsistent_dates.empty:
        print("\nInconsistências de datas encontradas:")
        num_inconsistent = len(inconsistent_dates)
        percentage_inconsistent = (num_inconsistent / len(df)) * 100
        print(f"  Quantidade de linhas com inconsistências: {num_inconsistent}")
        print(f"  Percentual sobre o total: {percentage_inconsistent:.2f}%")

        # Exportar o resultado para Excel
        output_file_name = "Inconsistencias_datas.xlsx"
        inconsistent_dates.to_excel(output_file_name, index=False)
        print(f"\nAs linhas com inconsistências de datas foram exportadas para: [1m{output_file_name}[0m")
        files.download(output_file_name)

    else:
        print("\nNenhuma inconsistência de datas encontrada.")
else:
    print("\nColunas 'Data de Entrega' ou 'Data de Lançamento' não encontradas no arquivo.")
```

```python
# --- Validação de Localização (Latitude/Longitude) ---
print("\n\n--- Validação de Localização (Latitude/Longitude) ---")
location_errors_df = pd.DataFrame()
if 'Latitude' in df.columns and 'Longitude' in df.columns:
    # Definir os limites de Porto Alegre (valores aproximados)
    POA_LAT_MIN = -30.2
    POA_LAT_MAX = -29.9
    POA_LON_MIN = -51.3
    POA_LON_MAX = -51.0

    df_location = df.copy()
    df_location['Erro_Localizacao'] = False

    # Verificar latitude/longitude nula ou não numérica
    for col in ["Latitude", "Longitude"]:
        df_location.loc[~pd.to_numeric(df_location[col], errors='coerce').notna(), "Erro_Localizacao"] = True

    # Converter para numérico após marcar os não numéricos como erros
    df_location["Latitude"] = pd.to_numeric(df_location["Latitude"], errors='coerce')
    df_location["Longitude"] = pd.to_numeric(df_location["Longitude"], errors='coerce')

    # Verificar se a latitude/longitude estão dentro  do limite de Porto Alegre
    df_location.loc[
        (df_location["Latitude"] < POA_LAT_MIN) |
        (df_location["Latitude"] > POA_LAT_MAX) |
        (df_location["Longitude"] < POA_LON_MIN) |
        (df_location["Longitude"] > POA_LON_MAX),
        "Erro_Localizacao"
    ] = True

    location_errors_df = df_location[df_location['Erro_Localizacao'] == True].copy()
    if not location_errors_df.empty:
        print(f"Foram encontradas {len(location_errors_df)} linhas com inconsistências de localização (Latitude/Longitude fora da região de Porto Alegre ou inválidas).")
        location_errors_df['Tipo_Erro'] = 'Localizacao_Inconsistente'
    else:
        print("Nenhuma inconsistência de localização (Latitude/Longitude) encontrada.")
else:
    print("\nColunas 'Latitude' ou 'Longitude' não encontradas no arquivo.")
```

# ***OPÇÃO PARA VALIDAR A LOCALIZAÇÃO SELECIONANDO AS CIDADES***

```python
# --- Validação de Localização (Latitude/Longitude) ---
print("\n\n--- Validação de Localização (Latitude/Longitude) ---")
# Initialize location_errors_df globally or in an accessible scope
location_errors_df_results = pd.DataFrame()

if 'Latitude' in df.columns and 'Longitude' in df.columns and 'Cidade' in df.columns:
    # Opções para o usuário
    modo_validacao_options = ['Validar todas as cidades da base', 'Escolher uma cidade específica']
    modo_validacao_dropdown = widgets.Dropdown(
        options=modo_validacao_options,
        value=modo_validacao_options[0],
        description='Modo de Validação:',
        disabled=False,
    )

    cidade_selecionada_dropdown = widgets.Dropdown(
        options=list(cidade_limites.keys()),
        description='Cidade:',
        disabled=True,
    )

    output_widget = widgets.Output()

    def on_modo_validacao_change(change):
        if change['new'] == 'Escolher uma cidade específica':
            cidade_selecionada_dropdown.disabled = False
        else:
            cidade_selecionada_dropdown.disabled = True

    modo_validacao_dropdown.observe(on_modo_validacao_change, names='value')

    display(modo_validacao_dropdown, cidade_selecionada_dropdown, output_widget)

    # Botão para iniciar a validação
    validar_button = widgets.Button(description="Iniciar Validação de Localização")
    display(validar_button)

    def on_validar_button_click(b):
        with output_widget:
            clear_output()
            current_location_errors = pd.DataFrame()

            if modo_validacao_dropdown.value == 'Validar todas as cidades da base':
                cidades_na_base = df['Cidade'].dropna().unique()
                for cidade in cidades_na_base:
                    if cidade in cidade_limites:
                        limites = cidade_limites[cidade]
                        df_cidade = df[df['Cidade'] == cidade].copy()
                        df_cidade['Erro_Localizacao'] = False

                        # Verificar latitude/longitude nula ou não numérica
                        for col in ["Latitude", "Longitude"]:
                            df_cidade.loc[~pd.to_numeric(df_cidade[col], errors='coerce').notna(), "Erro_Localizacao"] = True

                        df_cidade["Latitude"] = pd.to_numeric(df_cidade["Latitude"], errors='coerce')
                        df_cidade["Longitude"] = pd.to_numeric(df_cidade["Longitude"], errors='coerce')

                        # Verificar se a latitude/longitude estão dentro  do limite da cidade
                        df_cidade.loc[
                            (df_cidade["Latitude"] < limites["LAT_MIN"]) |
                            (df_cidade["Latitude"] > limites["LAT_MAX"]) |
                            (df_cidade["Longitude"] < limites["LON_MIN"]) |
                            (df_cidade["Longitude"] > limites["LON_MAX"]),
                            "Erro_Localizacao"
                        ] = True

                        erros_cidade = df_cidade[df_cidade['Erro_Localizacao'] == True].copy()
                        if not erros_cidade.empty:
                            erros_cidade['Tipo_Erro'] = f'Localizacao_Inconsistente_{cidade}'
                            current_location_errors = pd.concat([current_location_errors, erros_cidade])
                        print(f"Validação para {cidade} concluída.")
                    else:
                        print(f"Limites de Latitude/Longitude não definidos para {cidade}. Ignorando validação para esta cidade.")
            else: # Escolher uma cidade específica
                cidade = cidade_selecionada_dropdown.value
                if cidade in cidade_limites:
                    limites = cidade_limites[cidade]
                    df_cidade = df[df['Cidade'] == cidade].copy()
                    df_cidade['Erro_Localizacao'] = False

                    # Verificar latitude/longitude nula ou não numérica
                    for col in ["Latitude", "Longitude"]:
                        df_cidade.loc[~pd.to_numeric(df_cidade[col], errors='coerce').notna(), "Erro_Localizacao"] = True

                    df_cidade["Latitude"] = pd.to_numeric(df_cidade["Latitude"], errors='coerce')
                    df_cidade["Longitude"] = pd.to_numeric(df_cidade["Longitude"], errors='coerce')

                    # Verificar se a latitude/longitude estão dentro  do limite da cidade
                    df_cidade.loc[
                        (df_cidade["Latitude"] < limites["LAT_MIN"]) |
                        (df_cidade["Latitude"] > limites["LAT_MAX"]) |
                        (df_cidade["Longitude"] < limites["LON_MIN"]) |
                        (df_cidade["Longitude"] > limites["LON_MAX"]),
                        "Erro_Localizacao"
                    ] = True

                    erros_cidade = df_cidade[df_cidade['Erro_Localizacao'] == True].copy()
                    if not erros_cidade.empty:
                        erros_cidade['Tipo_Erro'] = f'Localizacao_Inconsistente_{cidade}'
                        current_location_errors = pd.concat([current_location_errors, erros_cidade])
                    print(f"Validação para {cidade} concluída.")
                else:
                    print(f"Limites de Latitude/Longitude não definidos para {cidade}. Não foi possível validar.")

            if not current_location_errors.empty:
                print(f"Foram encontradas {len(current_location_errors)} linhas com inconsistências de localização.")
                # Update the results variable
                global location_errors_df_results
                location_errors_df_results = pd.concat([location_errors_df_results, current_location_errors])
            else:
                print("Nenhuma inconsistência de localização encontrada para as cidades selecionadas.")


    validar_button.on_click(on_validar_button_click)

else:
    print("\nColunas 'Latitude', 'Longitude' ou 'Cidade' não encontradas no arquivo. Validação de localização não pode ser realizada.")
```
